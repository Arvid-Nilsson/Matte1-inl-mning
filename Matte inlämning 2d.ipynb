# Graf 2d

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 0.001
N = 25
g = 9.82
K_tot = 1
K = K_tot * (N-1)
m_tot = 1
m = m_tot/N
L_0tot = 1
L_0 = L_0tot / (1-N)
maxiter = 10000
```

```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

```python
X = np.linspace(0,1,N)
X[find_nearest(X,1/3)] = 1/3
X[find_nearest(X,2/3)] = 2/3
```

```python
Y = np.ones(len(X)) * 0.5
Y[0] = 0
Y[-1] = 0
Y[np.where(X == 1/3)] = 1
Y[np.where(X == 2/3)] = 1
```

```python
plt.plot(X, Y) 
```

```python
def calcForces(x1, y1, x2, y2, K_, L0, M, G):
    dx = x2 - x1
    dy = y2 - y1
    
    # Calculate the current distance between the two points
    L = np.sqrt(dx**2 + dy**2)
    
    if L == 0:
        return (0.0, M * -G)
    
    #Calculating spring forces
    Fx_spring = K_ * (L - L0) * (dx / L)
    Fy_spring = K_ * (L - L0) * (dy / L)
    
    # Calculating Gravitational force
    Fy_gravity = -M * g
    
    # Total forces on point 1
    Fx_total = Fx_spring
    Fy_total = Fy_spring + Fy_gravity
    
    return (Fx_total, Fy_total)
```

```python
maxF = np.zeros(len(X))
maxFPerIter = np.zeros(maxiter)

for j in range(maxiter):
    for i in range(0, len(X)):
        x = X[i]
        y = Y[i]

        if x not in [0, 1/3, 2/3, 1]:
            xl = X[i-1]
            yl = Y[i-1]
            xr = X[i+1]
            yr = Y[i+1]
            
            Fx_l, Fy_l = calcForces(x, y, xl, yl, K, L_0, m, g)
            Fx_r, Fy_r = calcForces(x, y, xr, yr, K, L_0, m, g)
            Fxtot = Fx_l + Fx_r
            Fytot = Fy_l + Fy_r
            
            x_new = x + a * Fxtot
            y_new = y + a * Fytot
            X[i] = x_new
            Y[i] = y_new
            
            Ftot = np.sqrt(Fytot**2 + Fxtot**2)
            maxF[i] = Ftot
    
    maxFPerIter[j] = maxF.max()
    maxF = np.zeros(len(X))
```
